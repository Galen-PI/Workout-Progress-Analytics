```python
#establishes new identifiers for table to begin creating multiple tables for refrence
col0 = df.iloc[:, 0]
metric_names = ['Total Load', 'Total Sets', 'Total Reps', 'Weights Per Rep']
muscles = col0[~col0.isna() & ~col0.str.strip().isin(metric_names)].str.strip().tolist()
```

```python
#begin total load table
muscles = ["Chest","Shoulders", "Triceps", "Upper Back", "Traps", "Biceps", "Quads", "Hamstrings", "Calves"] #mucle groups
weeks = df.columns[1:] #column 0 is for the year and weeks, so begining on column 1
total_load_table = pd.DataFrame(index=weeks, columns=muscles)
```

```python
for muscle in muscles:
    muscle_idx = df.index[df.iloc[:, 0] == muscle][0]
    muscle_rows = col0[col0.str.strip()== muscle]
    if len(muscle_rows) == 0:
      continue
    total_load_row = df.iloc[muscle_idx + 1, 1:]
    total_load_table[muscle] = total_load_row.values
```
