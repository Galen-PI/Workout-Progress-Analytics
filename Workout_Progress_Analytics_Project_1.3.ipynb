```python
#begin total load table
muscles = ["Chest","Shoulders", "Triceps", "Upper Back", "Traps", "Biceps", "Quads", "Hamstrings", "Calves"] #mucle groups
weeks = df.columns[1:] #column 0 is for the year and weeks, so begining on column 1
total_load_table = pd.DataFrame(index=weeks, columns=muscles)
```

```python
for i, muscle in enumerate(muscles):
  muscle_rows = col0[col0.str.strip()== muscle]
  if len(muscle_rows) == 0:
      continue
  muscle_idx = muscle_rows.index[0]
  if i < len(muscles) - 1:
    next_muscle_idx = col0[col0.str.strip() == muscles[i + 1]].index[0]
  else:
    next_muscle_idx = len(df)
  muscle_block = df.iloc[muscle_idx+1 : next_muscle_idx, :]
  mask = muscle_block.iloc[:,0].str.strip() == "Total Load"
  total_load_row = muscle_block[mask].iloc[0, 1:]
  total_load_table[muscle] = total_load_row.values.flatten()
```
